```python
import pandas as pd
import numpy as np
```

```python
students = pd.read_csv('student-math.csv')
```

```python
students.head()
```

```python
#Convert the target column from a grade percentage to a pass/fail
students['Pass_Fail'] = np.where(students['G3'] < 12, 0, 1)
```

```python
students.head()
```

```python
students.info()
```

```python
dummies_col = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 
               'reason', 'guardian', 'schoolsup', 'famsup', 'paid', 'activities',
              'nursery', 'higher', 'internet', 'romantic']
df_dummies = pd.get_dummies(students[dummies_col])
```

```python
students = students.drop(dummies_col, axis=1)

students = df_dummies.join(students)
students.head()
```

```python
students.info()
```

```python
X = students.drop(['G3', 'Pass_Fail'], axis=1)
y = students['Pass_Fail']
```

```python
from sklearn import cross_validation
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(
    X, y, test_size=0.2, random_state=50)
print(X_train.shape)
print(y_train.shape)
```

Data:  Student Performance Data Set
https://archive.ics.uci.edu/ml/datasets/student+performance

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

lr = LogisticRegression()
lr.fit(X_train, y_train)

predictions_1 = lr.predict(X_test)

print('Logistic Regression Classification')
print('Accuracy', accuracy_score(predictions_1, y_test))
print('Confusion Matrix\n', confusion_matrix(predictions_1, y_test))
print('Classification Matrix\n', classification_report(predictions_1, y_test))

```

```python
#Support Vector Machine
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train, y_train)

predictions_2 = svc.predict(X_test)

print('SVM Classification')
print('Accuracy', accuracy_score(predictions_2, y_test))
print('Confusion Matrix\n', confusion_matrix(predictions_2, y_test))
print('Classification Matrix\n', classification_report(predictions_2, y_test))

```

```python
#Decsion Tree
from sklearn import preprocessing, tree

dtree = tree.DecisionTreeClassifier(criterion='entropy')
dtree.fit(X_train, y_train)

predictions_3 = dtree.predict(X_test)

print('Decision Tree Classification')
print('Accuracy', accuracy_score(predictions_3, y_test))
print('Confusion Matrix\n', confusion_matrix(predictions_3, y_test))
print('Classification Matrix\n', classification_report(predictions_3, y_test))

```
